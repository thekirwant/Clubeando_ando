###Determinar el número de clústeres
Para determinar la manera en la que queremos que nuestros clústeres se organicen, podemos utilizar el método de la curva del codo, que consiste en la suma de errores cuadrado en cada número de clústeres para todos los datos que necesitemos, midiendo la distancia usando una métrica euclidiana; para esto, se busca en dónde se dibuja una curva o un "codo" y eso nos indica cuántos clústeres debemos usar.

```
arr = df.to_numpy() 

print ('\nNumpy Array\n----------\n', arr)
```

```
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import bokeh as bk 
from sklearn.cluster import KMeans
grupos = []
import numpy as np

K= range(1,10)
for k in K: 
	Modelokmean= KMeans(n_clusters=k).fit(arr)
	Modelokmean.fit(arr)
	grupos.append(sum(np.min(cdist(arr, Modelokmean.cluster_centers_,'euclidean'), axis = 1)) / df.shape[0])
plt.style.use('dark_background')
plt.plot(K,grupos,"bx-") 
plt.xlabel('Clústeres')
plt.ylabel('Suma Total de Errores Cuadráticos')
plt.title('Método del Codo')
plt.show() 
```

```
import plotly.express as px

K
for template in ["plotly_dark"]:
  fig = px.line(df, K , grupos , title='Método del codo', 
                template=template)
  fig.show()
```

### Clustering - K-means

```
import numpy as np
from sklearn.cluster import KMeans

np.random.seed(123)
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit(df)
existing_df_2d['cluster'] = pd.Series(clusters.labels_, index=existing_df_2d.index)

import matplotlib.pyplot as plt
pca = PCA(n_components=2)
pca.fit(df)

plt.style.use('dark_background')

import numpy as np
styles = ['#ED8E7A','r','darkred']
ax = existing_df_2d.plot(
    kind='scatter',
    x='PC2',y='PC1',
    c=[styles[i] for i in existing_df_2d.cluster.astype(np.int)],
    s=100,
    figsize=(16,8))
for i, country in enumerate(df.index):
    ax.annotate(country, (existing_df_2d.iloc[i].PC2, existing_df_2d.iloc[i].PC1))
```
